```python
## Oğuz Tanatar 05210000783
import matplotlib.pyplot as plt
import numpy as np

np.seterr(divide='ignore', invalid='ignore')
x = np.linspace(0,np.pi*2)
ab = 83
bc = 90
w = 783
#Sinüs kuralı kullanarak BC çubuğunun açısal hızını bulacağız.
# bc/sin(x) = ab/sin(a)
a = 90*np.arcsin(np.sin(x)/80)
aw = np.cos(x)/(83*np.cos(a))

#### Hızları bulmak için w*r formülünü kullanacağız. ####
vc = 90*(np.sin(x)+np.tan(a)*np.cos(x))
vb = 83*(np.cos(x)-np.sin(x))
### Konumlarını bulup grafiğe dökelim. ###
bx = 83*np.cos(x)
by = 83*np.sin(x)
cy = 0
cx = 90-(np.sin(x)+np.tan(a)*np.cos(x))

plt.title("Konum grafikleri")
plt.plot(x,bx)
plt.plot(x,cx)
plt.xlabel("Açı değerleri")
plt.ylabel("Konumlar")
plt.legend(["C noktasının konumu","B noktasının konumu"])
plt.show()




#Hız grafiklerini çizdirelim.

plt.title("Hız grafikleri")
plt.xlabel("Açı değerleri")
plt.ylabel("Hız değerleri")
plt.plot(x,vb)
plt.plot(x,vc)
plt.legend(["C noktasının hızı","B noktasının hızı"])
plt.show()
```
